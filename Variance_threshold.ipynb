```
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```
train=pd.read_csv('train_pca_0.99999999.csv')
test=pd.read_csv('test_pca_0.99999999.csv')
var=VarianceThreshold(0.000002)
var.fit(train.drop(['Culture','0.1','1.1','2.1'],axis=1))
```

```
new_train=pd.DataFrame(var.transform(train.drop(['Culture','0.1','1.1','2.1'],axis=1)))
new_train
```

```
new_test=pd.DataFrame(var.transform(test.drop(['0.1','1.1','2.1'],axis=1)))
```

```
new_train['Culture']=train['Culture']
```

```
new_train.to_csv('train_pca_var.csv',index=False)
new_test.to_csv('test_pca_var.csv',index=False)
```
