```python
import pandas as pd
```

```python
pred=pd.read_csv('preds/fastai0.csv')
for i in [1,6,7,15,42,70,80,96,99,154,482,666,854,123456,123465]:
    p1=pd.read_csv('preds/fastai'+str(i)+'.csv')
    pred=pd.concat([pred,p1],axis=1)
pred
```

```python
pred.columns=['0','1','6','7','15','42','70','80','96','99','154','482','666','854','123456','123465']
```

```python
pred
```

```python
final=[]
for i in range(len(pred.index)):
    final.append(pred.loc[i].value_counts().index.tolist()[0])
    print(pred.loc[i].value_counts())
```

```python
pred['final']=pd.Series(final)
pred
```

```python
pd.Series(pred['final']).to_csv('preds/final_fastai.csv', index=False, header=['Culture'])
```
