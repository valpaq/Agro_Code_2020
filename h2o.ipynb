```python
import h2o
import pandas as pd
h2o.init()
```

```python
train=h2o.import_file("data/train_dropped_290_without_nans.csv")[1:,:]
test=h2o.import_file('data/test_dropped_290_without_nans.csv')[1:,:]
y_train=h2o.import_file('data/y_train.csv')['Culture']
```

```python
y_train
```

```python
train.describe()
```

```python
test=test[1:,:]
test
```

```python
for i in ['C1','C2']:
    train[i]=train[i].asfactor()
    test[i]=test[i].asfactor()
train['Culture']=y_train.asfactor()
```

```python
from h2o.estimators import H2ODeepLearningEstimator
dl = H2ODeepLearningEstimator(distribution="multinomial",
                               hidden=[586,784],
                               epochs=1000,
                               train_samples_per_iteration=-1,
                               reproducible=True,
                               activation="Rectifier",
                               single_node_mode=False,
                               balance_classes=True,
                               force_load_balance=True,
                               seed=42,
                               score_training_samples=0,
                               score_validation_samples=0,
                               stopping_rounds=10)
dl.train(x=(train.drop('Culture',axis=1)).columns,
          y="Culture",
          training_frame=train)

# Eval performance:
dl.model_performance()
```

```python
pred = dl.predict(test)
```

```python
test_y = pred.as_data_frame()
```

```python
pd.Series(test_y['predict']).to_csv('preds/h2o_dl.csv', index=False, header=['Culture'])
```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.estimators.stackedensemble import H2OStackedEnsembleEstimator
from h2o.grid.grid_search import H2OGridSearch
nfolds = 5

# There are a few ways to assemble a list of models to stack together:
# 1. Train individual models and put them in a list
# 2. Train a grid of models
# 3. Train several grids of models
# Note: All base models must have the same cross-validation folds and
# the cross-validated predicted values must be kept.


# 1. Generate a 2-model ensemble (GBM + RF)

# Train and cross-validate a GBM
my_gbm = H2OGradientBoostingEstimator(distribution="multinomial",
                                      ntrees=1000,
                                      max_depth=5,
                                      min_rows=2,
                                      learn_rate=0.01,
                                      nfolds=nfolds,
                                      fold_assignment="Stratified",
                                      keep_cross_validation_predictions=True,
                                      seed=42)
my_gbm.train(x=(train.drop('Culture',axis=1)).columns,
          y="Culture",
          training_frame=train)


# Train and cross-validate a RF
my_rf = H2ORandomForestEstimator(ntrees=1000,
                                 nfolds=nfolds,
                                 fold_assignment="Stratified",
                                 keep_cross_validation_predictions=True,
                                 seed=1)
my_rf.train(x=(train.drop('Culture',axis=1)).columns,
          y="Culture",
          training_frame=train)


# Train a stacked ensemble using the GBM and GLM above
ensemble = H2OStackedEnsembleEstimator(model_id="my_ensemble_binomial",
                                       base_models=[my_gbm, my_rf, dl])
ensemble.train(x=(train.drop('Culture',axis=1)).columns,
          y="Culture",
          training_frame=train)

# Eval ensemble performance on the test data
perf_stack_test = ensemble.model_performance(test)

# Compare to base learner performance on the test set
perf_gbm_test = my_gbm.model_performance(test)
perf_rf_test = my_rf.model_performance(test)
baselearner_best_auc_test = max(perf_gbm_test.F1(), perf_rf_test.F1())
stack_auc_test = perf_stack_test.F1()
print("Best Base-learner Test AUC:  {0}".format(baselearner_best_auc_test))
print("Ensemble Test AUC:  {0}".format(stack_auc_test))

# Generate predictions on a test set (if neccessary)
pred = ensemble.predict(test)
```
